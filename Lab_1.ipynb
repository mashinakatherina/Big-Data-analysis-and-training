### Part 1

Get Data using API

```python
import requests
import json
import pandas as pd
import seaborn as sns
```

```python
#Swagger API https://www.albion-online-data.com/api/swagger/index.html
url = 'https://www.albion-online-data.com/api/v2/stats/prices/T4_BAG,T5_BAG'

response = requests.get(url)
json_data = json.loads(response.text)
with open('data.txt', 'w') as outfile:
    json.dump(json_data, outfile)
```

```python
with open("/content/data.txt") as f:
  d = json.load(f)
  df = pd.DataFrame.from_dict(d)
  print(df)
```

```python
df_filter = df ['city']. isin (['Black Market', 'Martlock Portal']) 
filtered_df = df [df_filter]
filtered_df
```

```python
pd.set_option('max_rows', 50)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
numeric_columns = ['sell_price_max', 'sell_price_min', 'quality']
```

```python
(filtered_df.sort_values(by=['item_id'])
 .head(50)
 .style
 .format('{:.1f}', na_rep='-')
 .format({'species': lambda x:x.lower(),
          'island': lambda x:x.lower(),
          'sex': lambda x: '-' if pd.isna(x) else x.lower()
         })
 .highlight_null(null_color='lightgrey')
 .highlight_max(color='yellowgreen', subset=numeric_columns)
 .highlight_min(color='coral', subset=numeric_columns)
)
```
